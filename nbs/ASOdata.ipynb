```python
import os
import matplotlib.pyplot as plt
import glob
import matplotlib as mpl
import xarray as xr
datapath = '/glade/u/home/mcowherd/ASO/'
```

```python
tif_files = glob.glob(datapath + '*.tif')
#for fn in tif_files:
#    print(f'gdal_translate -of NetCDF {fn} {fn[0:-4]}.nc')
for fn in tif_files:
    print(f'gdalwarp -s_srs EPSG:26913 -t_srs EPSG:4326 {fn} {fn[0:-4]}_rp.tif')
    print(f'gdal_translate -of NetCDF {fn[0:-4]}_rp.tif {fn[0:-4]}_rp.nc')
```

```python
nc_files50 = glob.glob(datapath + '*50m.nc')
nc_files3 = glob.glob(datapath + '*3m.nc')
```

```python
for file_path in nc_files50:
    data = xr.open_dataset(file_path)
    plot_data = data.to_array()[1].data.astype(float)
    fig, ax = plt.subplots()
    cax = ax.pcolormesh(plot_data, cmap='viridis')
    fig.colorbar(cax)
    ax.set_title(f"{os.path.basename(file_path)}")
    plt.show()
```
